# Gai/Gen: Text-to-Image (TTI)

## 1.1 Setting Up

We will create a seperate virtual environment for this to avoid conflicting dependencies that each underlying model requires.

```sh
conda create -n TTI python=3.10.10 -y
conda activate TTI
pip install ".[TTI]"
```

## 2 Examples

### 2.1 OpenAI DALL-E 3

````python
## 2.1 OpenAI DALL-E 3

print("GENERATING:")
from gai.gen import Gaigen
gen = Gaigen.GetInstance("./gai.json")
gen.load('openai-dalle3')
response = gen.create(
  messages="a white siamese cat"  
)
import requests
response = requests.get(response.data[0].url)
from PIL import Image
from io import BytesIO
img = Image.open(BytesIO(response.content))
img
````
